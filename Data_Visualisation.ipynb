```python
import matplotlib.pyplot as plt

# Selecting numeric columns
numeric_columns = [
    'Ajker Potrika credibility',
    'Desh Rupantor credibility',
    'Prothom Alo',
    'Daily Naya Diganta',
    'Dainik Amader Shomoy',
    'Jugantor',
    'Daily Inqilab',
    'SAMAKAL ',
    'Daily Janakantha',
    'The Daily Ittefaq',
    'Bhorer Kagoj',
    'Bangladesh Pratidin',
    'sangbad',
    'Jai Jai Din',
    'Mzamin',
    'The Daily Star',
    'kalerkantho',
    'The Financial Express',
    'The Daily Sangram',
    'Amardesh Online'
]

# Convert columns to numeric, coercing errors to NaN
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Replace NaN with 0 (optional, depending on how you want to handle missing data)
df[numeric_columns] = df[numeric_columns].fillna(0)

# Replace values less than 0 with 0 and values greater than 1 with 1
df[numeric_columns] = df[numeric_columns].clip(lower=0, upper=1)

# Plot cumulative sums
plt.figure(figsize=(14, 8))
for column in numeric_columns:
    cumulative_sum = df[column].cumsum()
    plt.plot(cumulative_sum, label=column)

#plt.title("Cumulative Sum of Adjusted Values by Column")
plt.xlabel("#No of Respondent")
plt.ylabel("Cumulative Sum of Newspaper Credibility Score")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title="Newspapers")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Selecting numeric columns
numeric_columns = [
    'Ajker Potrika credibility',
    'Desh Rupantor credibility',
    'Prothom Alo',
    'Daily Naya Diganta',
    'Dainik Amader Shomoy',
    'Jugantor',
    'Daily Inqilab',
    'SAMAKAL ',
    'Daily Janakantha',
    'The Daily Ittefaq',
    'Bhorer Kagoj',
    'Bangladesh Pratidin',
    'sangbad',
    'Jai Jai Din',
    'Mzamin',
    'The Daily Star',
    'kalerkantho',
    'The Financial Express',
    'The Daily Sangram',
    'Amardesh Online'
]

# Convert columns to numeric, coercing errors to NaN
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Replace NaN with 0 (optional, depending on how you want to handle missing data)
df[numeric_columns] = df[numeric_columns].fillna(0)

# Replace values less than 0 with 0 and values greater than 1 with 1
df[numeric_columns] = df[numeric_columns].clip(lower=0, upper=1)

# Generate a colormap with enough unique colors for all columns
colormap = cm.get_cmap('tab20', len(numeric_columns))  # 'tab20' provides 20 unique colors, scalable to more

# Plot cumulative sums with unique colors
plt.figure(figsize=(14, 8))
for i, column in enumerate(numeric_columns):
    cumulative_sum = df[column].cumsum()
    plt.plot(cumulative_sum, label=column, color=colormap(i))

plt.xlabel("# No. of Respondents")
plt.ylabel("Cumulative Sum of Newspaper Credibility Scores")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title="Newspapers")
plt.tight_layout()
plt.show()
```
